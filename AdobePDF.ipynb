# Chunking

```python
parser = AdobePDFParser(
    client_id=os.environ['ADOBE_CLIENT_ID'],
    client_secret=os.environ['ADOBE_CLIENT_SECRET'],
    mode='chunks', # mode in ["json", "chunks", "data"]
)
```

```python
for item in file_lst:
    print(item.split('/')[-1])
    loader = AdobePDFLoader(
        item,
        parser=parser
    )
    chunks = loader.load()
```
